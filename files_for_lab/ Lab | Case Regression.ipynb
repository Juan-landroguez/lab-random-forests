```python
data = pd.concat([numerical, categorical, targets], axis=1)
```

```python
#Select only TARGET_B=1, donnors
```

```python
data_b = data[data['TARGET_B'] == 1]
```

```python
X = data_b.drop(columns=['TARGET_B','TARGET_D'])
y = data_b['TARGET_D']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
num_cols = X.select_dtypes(np.number).columns.tolist()
cat_cols = X.select_dtypes(object).columns.tolist()

X_train_num = X_train[num_cols]
X_train_cat = X_train[cat_cols]

X_test_num = X_test[num_cols]
X_test_cat = X_test[cat_cols]
```

```python
#Scale the features either by using MinMax Scaler or a Standard Scaler. (train_num, test_num)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_num_scaled = scaler.fit_transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
```

```python
# Encode the categorical features using One-Hot Encoding or Ordinal Encoding. (train_cat, test_cat)
# fit only on train data, transform both train and test
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(drop='first')
X_train_cat_encoded = enc.fit_transform(X_train_cat).toarray()
X_test_cat_encoded = enc.transform(X_test_cat).toarray()
```

```python
# We concat again test and train X
```

```python
X_train = pd.concat([pd.DataFrame(X_train_num_scaled, columns=X_train_num.columns), pd.DataFrame(X_train_cat_encoded)], axis=1)
X_test = pd.concat([pd.DataFrame(X_test_num_scaled, columns=X_test_num.columns), pd.DataFrame(X_test_cat_encoded)], axis=1)
```

```python
#Fit a logistic regression model on the training data.
```

```python
#we train/fit our model like 
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
#Trying KNN to see if the score improves
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=10)

knn.fit(X_train, y_train)

Y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2 = knn.score(X_test, y_test)

print("Coeficiente de determinaciÃ³n RÂ²: {:.2f}".format(r2))
```

```python
# PCA
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=120)

X_train_pca = pca.fit_transform(X_train)

X_test_pca = pca.transform(X_test)
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X_train_pca,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train_pca)
r2_score(y_train, predictions)
```

```python
predictions_test = lm.predict(X_test_pca)
r2_score(y_test, predictions_test)
```

```python
#Random Forest
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

clf = RandomForestRegressor(n_estimators=120, # number of trees
                             max_depth=10, # maximum depth of the trees
                             max_samples=0.2, # fraction of X-train to use in each tree
                             random_state = 42)
clf.fit(X_train, y_train)

train_score = clf.score(X_train, y_train)
test_score = clf.score(X_test, y_test)
print(f"Training score: {train_score:.2f}")
print(f"Test score: {test_score:.2f}")

y_pred = clf.predict(X_test)
```
